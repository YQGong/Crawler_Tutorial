```python
def currentTimeStamp():
    return int(time.mktime(datetime.datetime.now().timetuple()))

def timeStampToTime(timestamp):
    return datetime.datetime.fromtimestamp(timestamp,pytz.timezone('Asia/Shanghai'))
```
